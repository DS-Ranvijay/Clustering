#Theoretical Questions

1. What is Unsupervised Learning in the context of Machine Learning?

Answer:

- Unsupervised learning is a type of machine learning in which the model learns patterns and relationships from unlabeled data. Unlike supervised learning, there are no predefined output labels. The goal is to discover hidden structures, clusters, or associations within the data.

- Example: Customer segmentation based on purchasing behavior.

2. How does the K-Means clustering algorithm work?

Answer:

- K-Means is a partition-based clustering algorithm that groups similar data points into K clusters.

- Steps:

  - Choose the number of clusters (K).
  - Randomly initialize K centroids.
  - Assign each data point to its nearest centroid.
  - Recalculate the centroid of each cluster.
  - Repeat steps 3 and 4 until the centroids no longer change or the maximum number of iterations is reached.

3. Explain the concept of a dendrogram in Hierarchical Clustering.

Answer:

- A dendrogram is a tree-like diagram used to visualize the process of hierarchical clustering. It shows how clusters are merged (agglomerative) or split (divisive) at different distance levels. By cutting the dendrogram at a chosen height, the desired number of clusters can be obtained.

4. What is the main difference between K-Means and Hierarchical Clustering?

- | K-Means                                         | Hierarchical Clustering                                       |
| ----------------------------------------------- | ------------------------------------------------------------- |
| Requires the number of clusters (K) beforehand. | Does not require specifying the number of clusters initially. |
| Uses centroids to form clusters.                | Builds a hierarchy of clusters.                               |
| Faster for large datasets.                      | Slower for large datasets.                                    |
| Produces flat clusters.                         | Produces a dendrogram representing nested clusters.           |


5. What are the advantages of DBSCAN over K-Means?

Answer:

- Does not require specifying the number of clusters.
- Can identify clusters of arbitrary shapes.
- Detects noise and outliers automatically.
- Performs better on datasets containing outliers.
- Works well for density-based clustering problems.

6. When would you use Silhouette Score in clustering?

Answer:

- Silhouette Score is used to evaluate the quality of clustering. It measures how similar a data point is to its own cluster compared to other clusters. It is commonly used to determine the optimal number of clusters.

  - Score close to 1: Well-separated clusters.
  - Score close to 0: Overlapping clusters.
  - Score below 0: Incorrect clustering.

7. What are the limitations of Hierarchical Clustering?

Answer:

- Computationally expensive for large datasets.
- Difficult to modify clusters once merged or split.
- Sensitive to noise and outliers.
- Requires significant memory.
- Choosing the correct linkage method can be challenging.

8. Why is feature scaling important in clustering algorithms like K-Means?

Answer:

- K-Means uses Euclidean distance to measure similarity. Features with larger scales dominate the distance calculation, leading to biased clusters. Feature scaling (e.g., Standardization or Min-Max Scaling) ensures that all features contribute equally.

9. How does DBSCAN identify noise points?

Answer:

- DBSCAN classifies points into:

  - Core points: Have at least MinPts neighbors within distance eps.
  - Border points: Close to a core point but have fewer than MinPts neighbors.
  - Noise points: Do not belong to any cluster because they are not within the neighborhood of any core point.

10. Define inertia in the context of K-Means.

Answer:

- Inertia is the sum of squared distances between each data point and its assigned cluster centroid. It measures how compact the clusters are.

11. What is the Elbow Method in K-Means clustering?

Answer:

- The Elbow Method helps determine the optimal number of clusters (K). It plots inertia against different values of K. The point where the decrease in inertia begins to slow significantly (forming an "elbow") is selected as the optimal K.

12. Describe the concept of "density" in DBSCAN.

Answer:

- In DBSCAN, density refers to the number of data points within a specified radius (eps). Areas with many nearby points are considered dense regions and form clusters, while sparse regions may be classified as noise.

13. Can Hierarchical Clustering be used on categorical data?

Answer:

- Yes. Hierarchical clustering can be applied to categorical data by using appropriate similarity or distance measures such as Hamming distance, Jaccard similarity, or Gower distance instead of Euclidean distance.

14. What does a negative Silhouette Score indicate?

Answer:

- A negative Silhouette Score indicates that data points are closer to neighboring clusters than to their own cluster. This suggests poor clustering or incorrect cluster assignments.

15. Explain the term "linkage criteria" in Hierarchical Clustering.

Answer:

- Linkage criteria determine how the distance between clusters is calculated during hierarchical clustering.

- Common linkage methods include:

  - Single Linkage: Minimum distance between clusters.
  - Complete Linkage: Maximum distance between clusters.
  - Average Linkage: Average distance between all pairs of points.
  - Ward's Linkage: Minimizes within-cluster variance.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer:

- K-Means assumes clusters are spherical, equally sized, and have similar densities. When clusters vary in size or density:

  - Large clusters may absorb smaller ones.
  - Dense clusters may be split incorrectly.
  - Sparse clusters may not be identified accurately.
  - Results become unreliable.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

Answer:

- DBSCAN has two main parameters:

  - eps (ε): Maximum distance between neighboring points.
  - MinPts: Minimum number of points required to form a dense region.

- Influence:

  - Larger eps creates larger clusters.
  - Smaller eps creates more noise points.
  - Higher MinPts requires denser clusters.
  - Lower MinPts allows smaller clusters to form.

18. How does K-Means++ improve upon standard K-Means initialization?

Answer:

- K-Means++ improves centroid initialization by selecting initial centroids that are far apart. This:

  - Reduces the chances of poor initial centroids.
  - Improves clustering accuracy.
  - Speeds up convergence.
  - Produces more consistent results compared to random initialization.

19. What is Agglomerative Clustering?

Answer:

- Agglomerative Clustering is a bottom-up hierarchical clustering method. Initially, each data point forms its own cluster. The algorithm repeatedly merges the two closest clusters until all points belong to a single cluster or a stopping criterion is reached.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer:

- While inertia measures cluster compactness, it always decreases as the number of clusters increases, making it unsuitable for comparing different values of K.

- Silhouette Score considers both:

  - Cluster cohesion (how close points are within the same cluster), and
  - Cluster separation (how far clusters are from each other).

- Therefore, Silhouette Score provides a more balanced and reliable evaluation of clustering quality than inertia alone.

21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='red', marker='X', s=200)
plt.title("K-Means Clustering")
plt.show()
```

24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X,_=make_blobs(n_samples=300, centers=4, random_state=42)

scores=[]

for k in range(2,6):
    labels=KMeans(n_clusters=k,random_state=42).fit_predict(X)
    scores.append(silhouette_score(X,labels))

plt.bar(range(2,6),scores)
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X,_=make_blobs(n_samples=300,centers=3,
               cluster_std=2.5,random_state=42)

kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(X)

x_min,x_max=X[:,0].min()-1,X[:,0].max()+1
y_min,y_max=X[:,1].min()-1,X[:,1].max()+1

xx,yy=np.meshgrid(np.arange(x_min,x_max,0.1),
                  np.arange(y_min,y_max,0.1))

Z=kmeans.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)

plt.contourf(xx,yy,Z,alpha=0.3)
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.show()
```

34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

X=load_breast_cancer().data

inertia=[]

for k in range(2,7):
    model=KMeans(n_clusters=k,random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)

plt.plot(range(2,7),inertia,marker='o')
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```
